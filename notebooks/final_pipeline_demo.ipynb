# Final Tennis Pipeline Demo

This notebook documents how to use the merged final pipeline for:

1. supervised action-recognition training
2. inference on a new raw tennis video
3. motion-analysis visualization


## Key fact

The current action-recognition model is **supervised** because it is trained from manually labeled tennis stroke windows.


```python
repo_root = '/media/cv/HDD/Throwing_Dataset/Tennis_Internship/Final_pipeline'
video_dir = '/media/cv/HDD/Throwing_Dataset/Tennis_Internship/videos'
excel_gt = '/media/cv/HDD/Throwing_Dataset/Tennis_Internship/Tennise_dataset_GT.xlsx'
pose_model = '/media/cv/HDD/Throwing_Dataset/Tennis_Internship/yolo11x-pose.pt'
checkpoint_dir = repo_root + '/checkpoints/action_recognition/current'
output_root = repo_root + '/outputs/demo_run'
```

## Train / evaluate the supervised action model

```python
train_cmd = f'''python {repo_root}/scripts/train_action_pipeline.py \
  --excel {excel_gt} \
  --video-dir {video_dir} \
  --pose-root /media/cv/HDD/Throwing_Dataset/Tennis_Internship/Project_pipeline/runs \
  --output-root {repo_root}/outputs/train_eval \
  --device cuda:0'''
print(train_cmd)
```

## Run the complete inference pipeline on a raw tennis video

```python
infer_cmd = f'''python {repo_root}/scripts/run_complete_pipeline.py \
  --video {video_dir}/tennis_video_2.mov \
  --pose-model {pose_model} \
  --checkpoint-dir {checkpoint_dir} \
  --output-root {output_root} \
  --target-person-id largest \
  --pose-device 0 \
  --handedness right \
  --h264'''
print(infer_cmd)
```

## Expected outputs

- pose CSV and tracked pose video
- predicted action segments CSV / JSON
- action recognition video
- motion-analysis video
- optional racket tracking outputs
